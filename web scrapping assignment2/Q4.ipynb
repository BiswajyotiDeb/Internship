```python
url='https://www.flipkart.com/'
driver.get(url)
```

```python
search_btn.click()
```

```python
# EXTRACTING PRODUCT NAME
p_name=driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
p_name[0:2]
```

```python
product=[]
for i in p_name:
    product.append(i.text)
product[0:2]
```

```python
# EXTRACTING PRODUCT PRICE
pr=driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
pr[0:2]
```

```python
price=[]
for i in pr:
    price.append(i.text)
price[0:2]
```

```python
# EXTRACTING PRODUCT DESCRIPTION
desc=driver.find_elements_by_xpath('//div[@class="_2B099V"]')
desc[0:2]
descrip=[]
for i in desc:
    descrip.append(i.text.strip().split('\n')[1])
descrip
```

```python
len(price),len(product),len(descrip)
```

```python
#creating a dataframe
pg1=pd.DataFrame()
pg1['Brand']=product
pg1['Price']=price
pg1['Description']=descrip
```
