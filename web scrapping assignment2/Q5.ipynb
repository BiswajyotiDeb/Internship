```python
#importing all required libraries
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#connect to web driver
driver=webdriver.Chrome(r'C:/Users/BISWAJYOTI/Downloads/chromedriver_win32/chromedriver.exe')
```

```python
url="https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-%20earpods-poweradapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKC%20TSVZAXUHGREPBFGI&marketplace."
driver.get(url)
```

```python
#clicking allreview using absolute xpath
all_review=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div/div/div[5]/div/a/div')
all_review.click()
```

```python
# EXTRACTING RATING
ratings=[]
for i in range(10):
    rating=driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]')
    for j in rating:
        ratings.append(j.text)
next_button=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]')
next_button.click()
```

```python
ratings
```

```python
# EXTRACTING REVIEW SUMMARY
review_summary=[]
for i in range(10):
    rs=driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
    for j in rs:
        review_summary.append(j.text)
next_button=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]')
next_button.click()
```

```python
review_summary
```

```python
# EXTRACTING review
review=[]
for i in range(10):
    rv=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
    for j in rv:
        review.append(j.text.replace('\n',' '))
next_button=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]')
next_button.click()
```

```python
review
```

```python
len(review),len(review_summary),len(ratings)
```

```python
#creating a dataframe
iphone_11=pd.DataFrame()
iphone_11['Rating']=ratings
iphone_11['Review Summary']=review_summary
iphone_11['Review']=review
```
