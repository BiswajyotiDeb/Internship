```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
page
```

```python
soup=bs(page.content)
print(soup.prettify())
```

```python
t=[]
for i in soup.find_all('tr',class_="rankings-block__banner"):
    t.append(i.text.replace(' ','').split())
t
```

```python
s=[]
for i in soup.find_all('tr',class_="table-body"):
     s.append(i.text.replace(' ','').split())
s
   
```

```python
ss=s[:9]
ss
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(t,columns=['Position','Team','ShortName','Matches','Points','Rating'])
df1
```

```python
df2=pd.DataFrame(ss,columns=['Position','Team','ShortName','Matches','Points','Rating'])
df2
```

```python
df3=df1.append(df2)
df3
```

```python
df3.set_index('Position',inplace=True)
df3
```

# 6)b)

```python
page1=requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
page1
```

```python
soup1=bs(page1.content)
print(soup1.prettify())
```

```python
b1=[]
for i in soup1.find_all('tr',class_="rankings-block__banner"):
    b1.append(i.text.replace(' ','').split())
b1
```

```python
bt1_=pd.DataFrame(b1,columns=['Rank','level','Player Name','Nation','Rating','Career Best'])
bt1_
```

```python
bt1_.drop('level',axis=1,inplace=True)
bt1_
```

```python
rank=[]
for i in soup1.find_all('span',class_="rankings-table__pos-number"):
    rank.append(i.text.replace(' ','').replace('\n',''))
rank
```

```python
bt=[]
for i in soup1.find_all('td',class_='table-body__cell rankings-table__name name'):
    bt.append(i.text.replace(' ','').replace('\n',''))
bt
```

```python
country=[]
for i in soup1.find_all('span',class_="table-body__logo-text"):
    country.append(i.text.replace(' ','').replace('\n',''))
country
```

```python
rating=[]
for i in soup1.find_all('td',class_="table-body__cell rating"):
    rating.append(i.text.replace(' ','').replace('\n',''))
rating
```

```python
career=[]
for i in soup1.find_all('td',class_="table-body__cell u-text-right u-hide-phablet"):
    career.append(i.text.replace(' ','').replace('\n',''))
career
```

```python
bl2_=pd.DataFrame({'Rank':rank,'Player Name':bt,'Nation':country,'Rating':rating,'Career Best':career})
bl2_=bl2_.head(9)
bl2_
```

```python
ranking=bt1_.append(bl2_)
ranking
```
