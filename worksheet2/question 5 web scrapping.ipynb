```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
s=[]
for i in soup.find_all('tr',class_="table-body"):
     s.append(i.text.replace(' ','').split())
s
   
```

```python
ss=s[:9]
ss
```

```python
df1=pd.DataFrame(t,columns=['Position','Team','ShortName','Matches','Points','Rating'])
df1
```

```python
page2=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
page2
```

```python
soup2=bs(page2.content)
print(soup2.prettify())
```

```python
bl1=[]
for i in soup2.find_all('tr',class_="rankings-block__banner"):
    bl1.append(i.text.replace(' ','').split())
bl1
```

```python
bl1_=pd.DataFrame(bl1,columns=['Rank','level','Player Name','Nation','Rating','Career Best'])
bl1_
```

```python
bl1_.drop('level',axis=1,inplace=True)
bl1_
```

```python
rank=[]
for i in soup2.find_all('span',class_="rankings-table__pos-number"):
    rank.append(i.text.replace(' ','').replace('\n',''))
rank
```

```python
player=[]
for i in soup2.find_all('td',class_='table-body__cell rankings-table__name name'):
    player.append(i.text.replace(' ','').replace('\n',''))
player
```

```python
country=[]
for i in soup2.find_all('span',class_="table-body__logo-text"):
    country.append(i.text.replace(' ','').replace('\n',''))
country
```

```python
rating=[]
for i in soup2.find_all('td',class_="table-body__cell rating"):
    rating.append(i.text.replace(' ','').replace('\n',''))
rating
```

```python
career=[]
for i in soup2.find_all('td',class_="table-body__cell u-text-right u-hide-phablet"):
    career.append(i.text.replace(' ','').replace('\n',''))
career
```

```python
bl2_=pd.DataFrame({'Rank':rank,'Player Name':player,'Nation':country,'Rating':rating,'Career Best':career})
bl2_=bl2_.head(9)
bl2_
```
